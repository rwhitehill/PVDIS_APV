## variables

```python
up,dp,sp,cp=sy.symbols('u^+,d^+,s^+,c^+')
um,dm,sm,cm=sy.symbols('u^-,d^-,s^-,c^-')
s2w=sy.Symbol('s_w^2')

eu=+2/sy.S(3)
ed=-1/sy.S(3)


gVu=+1/sy.S(2)-2*eu*s2w
gVd=-1/sy.S(2)-2*ed*s2w
gAu=+1/sy.S(2)
gAd=-1/sy.S(2)

gVe = -1/sy.S(2)+2*s2w
gAe = -1/sy.S(2)

C1u,C1d=sy.symbols('C1_u,C1_d')
C2u,C2d=sy.symbols('C2_u,C2_d')
Rc,Rs,Rv=sy.symbols('R_c,R_s,R_v')
```

```python
F1gg=1/sy.S(2)*(eu**2*(up+cp)+ed**2*(dp+sp))
F1gg
```

```python
F1gZ=eu*gVu*(up+cp)+ed*gVd*(dp+sp)
F1gZ
```

```python
F3gZ=2*(eu*gAu*(um+cm)+ed*gAd*(dm+sm))
F3gZ
```

## $a_1$ version 1

```python
a1=(2*gAe*(F1gZ/F1gg)).simplify()
a1
```

```python
a1.subs(cp,0).subs(sp,0).subs(dp,up).simplify()
```

## $a_1$ version 2

```python
a1=2*(C1u*eu*(up+cp)+C1d*ed*(dp+sp))/(eu**2*(up+cp)+ed**2*(dp+sp))
a1=a1.subs(C1u,2*gAe*gVu).subs(C1d,2*gAe*gVd)
a1=a1.simplify()
a1
```

## $a_1$ version 3

```python
num=C1u*eu*((up+dp)/2+cp)+C1d*ed*((up+dp)/2+sp)
den=eu**2*((up+dp)/2+cp)+ed**2*((up+dp)/2+sp)
a1=2*num/den


a1=a1.subs(cp,Rc/2*(up+dp))
a1=a1.subs(sp,Rs/2*(up+dp))

a1=a1.simplify()
a1
```

## $a_3$ version 1

```python
a3=(gVe*(F3gZ/F1gg)).simplify()
a3
```

## $a_3$ version 2

```python
a3=2*(C2u*eu*(um+cm)+C2d*ed*(dm+sm))/(eu**2*(up+cp)+ed**2*(dp+sp))
a3=a3.subs(C2u,2*gVe*gAu).subs(C2d,2*gVe*gAd)
a3=a3.simplify()
a3
```

## $a_3$ version 3

```python
num=C2u*eu*((um+dm)/2+cm)+C2d*ed*((um+dm)/2+sm)
den=eu**2*((up+dp)/2+cp)+ed**2*((up+dp)/2+sp)
a3=2*num/den
a3=a3.subs(cp,Rc/2*(up+dp))
a3=a3.subs(sp,Rs/2*(up+dp))
a3=a3.subs(cm,0)  #!<--ATTENTION
a3=a3.subs(sm,0)  #!<--ATTENTION
a3=a3.subs(um,Rv*(up+dp)-dm)
a3=a3.simplify()
a3
```

## $a_3$ version 4

```python
num=C2u*eu*((um+dm)/2+cm)+C2d*ed*((um+dm)/2+sm)
den=eu**2*((up+dp)/2+cp)+ed**2*((up+dp)/2+sp)
a3=2*num/den
a3=a3.subs(cp,Rc/2*(up+dp))
a3=a3.subs(sp,Rs/2*(up+dp))
a3=a3.subs(cm,0)
#a3=a3.subs(sm,0)
delta=sy.Symbol('Delta')
a3=a3.subs(um,Rv*(up+dp)-dm)
a3=a3.subs(sm,delta*(up+dp))
a3=a3.simplify()
a3
```

## Apv 

```python
Y1,Y3=sy.symbols('Y_1,Y_3')
```

```python
Apv_red=+1/sy.S(4)*(a1*Y1+a3*Y3) * (2/sy.S(3))
Apv_red=Apv_red.subs(delta,0)
Apv_red.simplify()
```

```python
Apv_red=+1/sy.S(4)*(a1*Y1+a3*Y3) * (2/sy.S(3))
Apv_red.simplify()
```

```python
a1=(2*gAe*(F1gZ/F1gg)).simplify()
a1
```

```python
a3=(gVe*(F3gZ/F1gg)).simplify()
a3
```

```python
Apv,GF,Q,alpha,Y1,Y3=sy.symbols(r'A,G_F,Q,\alpha,Y1,Y3')
_up,_dp,_sp,_cp,_um,_dm,_sm,_cm=sy.symbols(r'up,dp,sp,cp,um,dm,sm,cm')
```

```python
expr=a1*Y1+a3*Y3
expr
```

```python
sol=sy.solve([Apv-expr],s2w)
```

```python
sol_=sol[s2w].simplify()#.replace(up,_up).replace(dp,_dp).replace(cp,_cp).replace(sp,_sp)
#sol_=sol_.simplify().replace(um,_um).replace(dm,_dm).replace(cm,_cm).replace(sm,_sm)

sol_=sol_.replace(Y1,1).replace(Y3,1)
sol_=sol_.replace(cp,0).replace(cm,0)
sol_=sol_.replace(sp,0).replace(sm,0)
sol_=sol_.replace(dp,(_up+_dp)/2)
sol_=sol_.replace(up,(_up+_dp)/2)
sol_=sol_.replace(_up,_dp)
sol_=sol_.replace(dm,(_um+_dm)/2)
sol_=sol_.replace(um,(_um+_dm)/2)
sol_=sol_.replace(_um,_dm)
sol_=sol_.replace(_dp,_dm)

sol_.simplify()
```
