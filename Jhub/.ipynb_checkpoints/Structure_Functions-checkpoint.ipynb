$ \displaystyle
F_{1}^{[\gamma,\gamma ùëû[0,2ùëíùëûùëîùëûùê¥,2ùëîùëûùëâùëîùëûùê¥]ùëì‚àíùëû(ùë•,ùëÑ2)Z, Z]}(x,Q^2) = \frac{1}{2} \sum_{q} [e_{q}^2, 2e_q g_V^{q}, (g_V^{q})^2 + (g_A^{q})^2] f_{q}^{+}(x,Q^2)
$

$ \displaystyle
F_{2}^{i}(x,Q^2) = 2xF_{1}^{i}(x,Q^2
$

$ \displaystyle
F_{3}^{[\gamma,\gamma Z, Z]}(x,Q^2) = \sum_{q} [0, 2e_q g_A^{q}, 2 g_V^{q} g_A^{q}] f_{q}^{-}(x,Q^2)
$

Note: 
- $f^{+}_{q} = f_{q} + f_{\bar{q}}$ and $f^{-}_{q} = f_{q} - f_{\bar{q}}$
- $g_V^{e} = -\frac{1}{2} + 2\sin^2{\theta_{w}} \quad g_A^{e} = -\frac{1}{2}$
- $g_V^{[u,c,t]} = \frac{1}{2} - \frac{4}{3}\sin^2{\theta_{w}} \quad g_A^{u,c,t} = \frac{1}{2}$
- $g_V^{[d,s,b]} = -\frac{1}{2} + \frac{2}{3}\sin^2{\theta_{w}} \quad g_A^{[d,s,b]} = -\frac{1}{2}$
- $\sin^2{\theta_{w}} = 0.23122$

```python
x  = np.logspace(-2,np.log10(0.7))
Q2 = 10.0

flavs = ['bb','cb','sb','ub','db','g','d','u','s','c','b']
iflavs = {flavs[i+5]: i for i in range(-5,6)}

xfs = {flavs[i+5]: [] for i in range(-5,6)}
for irep in range(len(JAM[:100])):
    for flav in flavs:
        xfs[flav].append([JAM[irep].xfxQ2(iflavs[flav],_,Q2) for _ in x])

for flav in flavs:
    xfs[flav] = np.array(xfs[flav])
    
q_plus  = {flavs[i+5]: (xfs[flavs[i+5]] + xfs[flavs[5-i]])/x for i in range(1,6)}
q_minus = {flavs[i+5]: (xfs[flavs[i+5]] - xfs[flavs[5-i]])/x for i in range(1,6)}
```

```python
eU  = 2.0/3.0
eD  = -1.0/3.0
eq  = {'d': eD, 'u': eU, 's': eD, 'c': eU, 'b': eD}

sin2_thw = 0.23122
gA = {flav: np.sign(eq[flav])/2.0 for flav in ['d','u','s','c','b']}
gV = {flav: gA[flav] - 2.0*eq[flav]*sin2_thw for flav in ['d','u','s','c','b']}
```

```python
shape = np.shape(q_plus['d'])

F1_g   = np.zeros(shape)
F1_gZ  = np.zeros(shape)
F1_Z   = np.zeros(shape)

F3_g   = np.zeros(shape)
F3_gZ  = np.zeros(shape)
F3_Z   = np.zeros(shape)

for flav in ['d','u','s','c','b']:
    F1_g  += 0.5*eq[flav]**2.0*q_plus[flav]
    F1_gZ += eq[flav]*gV[flav]*q_plus[flav]
    F1_Z  += 0.5*(gV[flav]**2.0+gA[flav]**2.0)*q_plus[flav]

    F3_gZ += 2.0*eq[flav]*gA[flav]*q_minus[flav]
    F3_Z  += 2.0*gV[flav]*gA[flav]*q_minus[flav]
```

```python
nrows,ncols = 1,1
fig, ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,5*nrows))

ax.plot(x,np.mean(F1_g,axis=0) ,'k-',label=r'$F_{1}^{\gamma}$')
ax.plot(x,np.mean(F1_gZ,axis=0),'k--',label=r'$F_{1}^{\gamma Z}$')
ax.plot(x,np.mean(F1_Z,axis=0) ,'k.',label=r'$F_{1}^{Z}$')

ax.plot(x,np.mean(F3_g,axis=0) ,'r-',label=r'$F_{3}^{\gamma}$')
ax.plot(x,np.mean(F3_gZ,axis=0),'r--',label=r'$F_{3}^{\gamma Z}$')
ax.plot(x,np.mean(F3_Z,axis=0) ,'r.',label=r'$F_{3}^{Z}$')
    
ax.legend(fontsize=20,ncols=2,columnspacing=0.2)
ax.semilogx()
ax.set_xlabel(r'$x$',size=20)
ax.tick_params(which='major',axis='both',direction='in',labelsize=20)
plt.show()
```

```python
nrows,ncols = 1,1
fig, ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,5*nrows))

ax.plot(x,np.mean(F1_g/F1_g,axis=0) ,'k-',label=r'$F_{1}^{\gamma}$')
ax.plot(x,np.mean(F1_gZ/F1_g,axis=0),'k--',label=r'$F_{1}^{\gamma Z}$')
ax.plot(x,np.mean(F1_Z/F1_g,axis=0) ,'k.',label=r'$F_{1}^{Z}$')

ax.plot(x,np.mean(F3_g/F1_g,axis=0) ,'r-',label=r'$F_{3}^{\gamma}$')
ax.plot(x,np.mean(F3_gZ/F1_g,axis=0),'r--',label=r'$F_{3}^{\gamma Z}$')
ax.plot(x,np.mean(F3_Z/F1_g,axis=0) ,'r.',label=r'$F_{3}^{Z}$')
    
ax.legend(fontsize=20,ncols=2,columnspacing=0.2)
ax.semilogx()
ax.set_xlabel(r'$x$',size=20)
ax.tick_params(which='major',axis='both',direction='in',labelsize=20)
plt.show()
```

# Comparing terms

```python
GF  = 1.1663787e-5
gAe = -0.5
gVe = -0.5 + 2.0*sin2_thw

alfa = 1.0/137.0

# JLab
M = 0.938
E = 12.0
root_s = np.sqrt(M**2.0 + 2.0*M*E)

# EIC
# root_s = 140.0

s = root_s**2.0
y = Q2/(s-M**2.0)/x
```

```python
term1 = 2.0*np.sqrt(2.0)*GF*Q2*gVe*(y**2.0-2.0*y+2.0)*F1_gZ
term3 = -np.sqrt(2)*GF*Q2*gAe*(y**2.0 - 2.0*y)*F3_gZ
term_g = 8.0*np.pi*alfa*(y**2.0-2.0*y+2.0)*F1_g
```

```python
nrows,ncols = 1,1
fig, ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,5*nrows))

ax.plot(x,np.mean(term1/term_g,axis=0) ,'k-')
ax.plot(x,np.mean(term3/term_g,axis=0),'k--')
    
ax.semilogx()
ax.set_xlabel(r'$x$',size=20)
ax.tick_params(which='major',axis='both',direction='in',labelsize=20)
plt.show()
```
